In the rapidly evolving environment of modern electronics manufacturing, the production of Printed Circuit Boards (PCBs) requires the highest precision, thus automated quality assurance systems are a critical bottleneck for operational efficiency. Traditional manual inspection techniques are very susceptible to human fatigue and subjective error, thus the semiconductor and assembly industries rely heavily on Automated Optical Inspection (AOI) systems based on high-resolution computer vision algorithms. The sampled PCB pixel data was processed by a series of descriptive and inferential statistical methods to extract meaningful engineering insights. The raw database retrieval was coupled with Python-based numerical computing. The pipeline starts with an in-memory SQLite connection to isolate our target table, running a query to pull the random sample, and loading the resulting schema directly into a Pandas DataFrame. The central part of our mathematical analysis is devoted to the central tendency of the normalized Red ($R$) channel for which we compute the sample mean, median and mode. The sample mean, defined by:$$\mu = \frac{1}{n} \sum_{i=1}^{n} x_i$$gives an approximate value of $0.584742$, which is the physical center of gravity for the red spectral distribution. This arithmetic mean is compared with the median ($0.556863$), which is the 50th percentile value of a sorted array, and the mode ($0.490196$), which is the most frequent digital intensity. The clear difference between these three metrics tells us immediately that we have an asymmetrical, non-symmetrical distribution. To quantify the physical spread and variability of the pixel colors we calculated the sample variance . $s^2 approx 0.047706$. The positive square root of the variance is the standard deviation $s approx 0.218418$. The standard deviation is very useful . It tells us that the average color intensity of a component of the board is about $ 21.84\% $ away from the mean . These descriptive metrics describe the immediate sample but we also use inferential statistics to understand the accuracy of our estimation in relation to the parent population of 723,000 pixels. This is done by calculating the Standard Error of the Mean (SEM) which is mathematically defined as:$$SEM = \frac{s}{\sqrt{n}}$$which results in $0.021842$. The SEM is an important indicator of statistical accuracy. It says that if we took many random samples of 100 pixels from the master circuit board image, the sample means would vary from the true, absolute population mean by only a standard amount of 2.18%. This shows that our sampling method is sound.

The empirical statistical analysis, and the following spatiotemporal distribution curve of the normalized Red ($R$) channel color spectrum, reveal the essential structural features of the examined Printed Circuit Boards. The density plot exhibits a strong bimodal distribution that corresponds both visually and mathematically to the physical and optical properties of the physical board components. The first distinct peak is tightly bunched together around the values $0.490196$ to $0.556863$ which represents the matte, desaturated grey test-pad surfaces which reflect lower levels of red light. The second peak with higher intensity from $0.713725$ to $0.768627$ corresponds to the highly reflective elements of copper traces, solder masks or lighter substrate elements covering the background of the circuit board. The dataset exhibits a pronounced positive skewness as evidenced by the computed sample mean ($\mu \approx 0.5847$) being displaced to the right of the median ($M \approx 0.5569$). The mismatch between the overlay of our theoretical red normal curve and the actual blue density bars further proves the fact that the dataset does not follow a standard Gaussian normal distribution which is mathematically proved by the skewness. From an engineering and machine learning perspective, this non-normality is very important; it warns system designers that classical parametric classification algorithms, designed to work on a clean and single-peaked bell curve, will perform badly in terms of high error rates when used for automated defect detection on these boards. Rather, the bimodal separation between the test pads and the background elements is distinctive and suggests that a non-parametric approach such as a local support vector machine, a decision tree or a binarized logistic regression would be far more effective at isolating target components.
